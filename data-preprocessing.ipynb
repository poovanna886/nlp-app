```python
import pandas as pd
import numpy as np
```

```python
brand=pd.read_csv('brand_category.csv')
brand=brand.drop('RECEIPTS',axis=1)
```

```python
np.shape(brand)
```

```python
brand.head()
```

```python
# renamed the brand_belongs_to_category to product_category as both conveyed the same information .
brand = brand.rename(columns={'BRAND_BELONGS_TO_CATEGORY': 'PRODUCT_CATEGORY'})
```

```python
brand.head()
```

```python
# Total number of product categories in brand_category data
len(np.unique(brand['PRODUCT_CATEGORY']))
```

```python
categories=pd.read_csv('categories.csv')
# dropped the category_id from the categories
categories=categories.drop('CATEGORY_ID',axis=1)
categories.head()
```

```python
np.shape(categories)
```

```python
retailer=pd.read_csv('offer_retailer.csv')
retailer.head()
```

```python
np.shape(retailer)
```

```python
result = pd.merge(retailer, brand, on='BRAND', how='outer')
result.head()
```

```python
result = pd.merge(result,categories, on='PRODUCT_CATEGORY', how='outer')
result.head()
```

```python
np.shape(result)
```

```python
# Check for null values
np.sum(result.isna())
# Retailer has 423 null values which can be removed from the dataset
```

```python
result= result.dropna()
```

```python
np.sum(result.isna())
```

```python
np.shape(result)
```

```python
result.head()
```

```python
# save the result data as 'final.csv' and this data for model training
result.to_csv('final.csv', index=False)
```
